## Control points in the Lab space

```python
pts_lab = [
    (20, 57, -77),
    (40, -9, -25),
    (60, -61, 59),
    (80, 10, 75),
]
```

```python
plot_lab_pts(pts_lab, markersize=15);
```

```python
def interp_lab(pts_lab, n):
    # unpack Lab channels
    l,a,b = np.asarray(pts_lab).T
    # Interpolate each channel linearly
    xp = np.linspace(0,1, len(l))
    x = np.linspace(0,1, 300)
    l_interp, a_interp, b_interp = [
        np.interp(x, xp, channel) for channel in [l,a,b]
    ]
    # repack the Lab channels
    pts_lab_interp = np.vstack((l_interp, a_interp, b_interp)).T
    return pts_lab_interp

pts_lab_interp = interp_lab(pts_lab, 50)
```

```python
plot_lab_pts(pts_lab_interp, markersize=20, markeredgewidth=0);
```

```python
def cmap_from_lab_pts(pts_lab):
    'Creates a colormap from the interpolation (in Lab space) of `pts_lab`'
    pts_lab_interp = interp_lab(pts_lab, 100)
    # compute the sRGB color
    pts_rgb = lab2rgb_list(pts_lab_interp)
    cmap = mpl.colors.LinearSegmentedColormap.from_list('Lab interp', pts_rgb, 100)
    return cmap
```

```python
cmap = cmap_from_lab_pts(pts_lab)
cmap(1.0)
```

## Test function for colormap plots

```python
def tfun_3hills(x,y):
    '''2D function on [-1,1]x[-1,1], with 3 hills of different heights,
    and one wide hole
    '''
    def bell(x,y):
        return np.exp(-(x**2+y**2))
    z = 0
    z -= bell(x,y)
    z += bell((x-0.7)*4, y*4)
    z += bell(x*4, (y-0.7)*4)*2
    z += bell((x+0.7)*4, y*4)*3
    return z
```

```python
x = np.linspace(-1,1, 300)
plt.plot(x, tfun_3hills(x, 0));
plt.title('cut along the x axis, at y=0')
plt.xlabel('x');
```

```python
def plot_tfun(tfun, cmap):
    y,x = np.ogrid[-1:1:100j, -1:1:100j]
    z = tfun(x,y)
    fig, ax = plt.subplots(2,2, figsize=(10,8))
    
    cmaps = [[cmap, 'YlGnBu_r'],
             ['gray', 'jet']]
    
    for i in [0,1]:
        for j in [0,1]:    
            im = ax[i,j].imshow(z, origin='lower', cmap=cmaps[i][j], extent=[-1,1, -1, 1]);
            plt.colorbar(im, ax=ax[i,j])
            if i == 0 and j == 0:
                title = 'cmap under test'
            else:
                title='cmap "{}"'.format(cmaps[i][j])
            ax[i,j].set_title(title)   
    
    fig.tight_layout()
```

```python
plot_tfun(tfun_3hills, cmap)
```
